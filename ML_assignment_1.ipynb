```
#ML-Assignment 1-1
import numpy as np
import matplotlib.pyplot as plt

#선형 함수
m = 20
x=[i for i in range(m)]
y=[4*b +8 for b in range(m)]
random = np.random.randn(m)

randx_point=[i for i in range(m)]
randy_point=[4*b +8 + random[b] for b in range(m)]



plt.scatter(randx_point,randy_point,color='black')
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('ML-Assignment1-1')
plt.legend(['line','dot'])
plt.show()




```

과제 1-1 :  
Y=4X + 8이라는 임의의 직선함수를 정하고
1의간격으로 그 직선을 중심으로 가우시안 표준 정규분포의 랜덤한 점 20개를 생성.  
위의 그래프는 파란선이 직선함수이고 검은 점들이 생성된 20개의 랜덤한 점들.


```
#ML-Assignment 1-3


rand_points = []

for i in range(m) :
  rand_points.append([randx_point[i],randy_point[i]])



theta0 =1
theta1 = 1
thetas = []
count = 10000
counts =[i for i in range(count)]

learningrate = 0.01
#Gradient descent 구현
for i in range(count) :
  temp = 0
  temp1 = 0
  for i in range (m) :
    temp += rand_points[i][0] * theta1 + theta0 - rand_points[i][1]
    temp1 += (rand_points[i][0] * theta1 + theta0 - rand_points[i][1]) * rand_points[i][0]
  temp = (temp*learningrate)/m
  temp1 = (temp1*learningrate)/m

  theta0 = theta0 - temp
  theta1 = theta1 - temp1
  thetas.append([theta0,theta1])

  #구현된 Gradient descent를 이용하여 Objectiv Function을 구현

J_thetas = []
for i in range(len(thetas)) :
  temp = 0
  
  for j in range(m) :
    temp += (thetas[i][0] +thetas[i][1]*rand_points[j][0] - rand_points[j][1])**2
  temp = temp/(2*m)
  J_thetas.append(temp)


plt.plot(counts , J_thetas)

print('Most Smallest theta 0 :',thetas[J_thetas.index(min(J_thetas))][0])
print('Most Smallest theta 1 :', thetas[J_thetas.index(min(J_thetas))][1])




plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('ML-Assignment1-3')
plt.legend(['J_theta'])
plt.show()

theta0value = thetas[J_thetas.index(min(J_thetas))][0]
theta1value = thetas[J_thetas.index(min(J_thetas))][1]
```

과제 1-3 :  
 세타0와 세타1의 초기값을 각각 1로 하고   
θ0(t+1)​:=θ0(t)​−αm1​∑i=1m​(hθ​(x(i))−y(i))  
θ1(t+1):=θ1(t)−α1m∑i=1m(hθ(x(i))−y(i))x(i)  
의 식에 따라서 Gradient descent 알고리즘을 이중포문을 이용하여 구현.

학습 횟수는 1만회 러닝레이트는 0.01로 설정하였으며 학습횟수와 러닝레이트에 따라 그래프의 형태와 값이 변경됨을 확인  

아래의 그래프는 Objective function의 학습 횟수에 따른 값을 나타내고 그 값이 수렴하는 것을 확인  

Gradient Descent 알고리즘에 의해 EnergyValue의 값이 최소가 되는 세타0와 세타1의 값을 print문으로 출력함  


```
#ML-Assignment 1-4

theta0s = []
theta1s = []

for i in range(len(thetas)) :

  theta0s.append(thetas[i][0])
  theta1s.append(thetas[i][1])




plt.plot(counts, theta0s,color= 'red')
plt.plot(counts, theta1s,color='blue' )



plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('ML-Assignment1-4')
plt.legend(['theta0','theta1'])
plt.show()
```
